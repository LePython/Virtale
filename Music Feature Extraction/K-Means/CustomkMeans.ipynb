```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python


from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);


```

```python


from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

```

```python


plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);

```

```python
#K-Means clustering with my own algorithm

cluster_center = np.array( [[0,1], [8,1]] )
print("Cluster Center: ",np.transpose(cluster_center)[1])

a = X[7]
b = cluster_center[1]

print("Arraygre von analysierten Daten: ", X.shape)

print("Ein beliebiger Punkt aus dem Datenarray: ",a)

abstand = np.linalg.norm(a-b)
print("Abstand:" ,abstand)



#for point in X:
#    print(point)
```
